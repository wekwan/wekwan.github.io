```python
import pandas as pd
import numpy as np
```

```python
daily_sales = pd.read_csv('daily_sales.csv')
daily_sales.head()
```

```python
new_cols_1 = [i.replace('-', '_') for i in daily_sales.columns.tolist()]
```

```python
daily_sales.columns = new_cols_1
```

```python
for i in daily_sales.columns:
    print(i)
    print(daily_sales[i].tolist())
    print()
```
